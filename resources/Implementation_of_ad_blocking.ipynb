```python
import pandas as pd
df= pd.read_csv("ad.data", header = None)
```

```python
df.head()
```

```python
df.rename(columns = {1558: "label"}, inplace=True)
```

```python
df.head()
```

```python
improper_rows = []
for index, row in df.iterrows():
    for col in df.columns:
        val = str(row[col]).strip()
        if val == "?":
            improper_rows.append(index)
```

1. `improper_rows = []` : Initializing an empty list named `improper_rows`. This is to store the indices of the rows in the DataFrame where some column has an empty string (""), hence deemed as being "improper".

2. `for index, row in df.iterrows():` : It is looping over the rows in the DataFrame. At each iteration, it gets the index of the row and the data of the row itself.3. `for col in df.columns:` : This second loop is nested inside the first loop. It is looping over each column in the DataFrame for a particular row.

4. `val = str(row[col]).strip()` : It is extracting the value in the current column and current row, converting it into a string (regardless of its initial data type), and then removing any leading or trailing white spaces from the string.

 5. `if val == "?":` : This line is checking whether our stripped string is an empty string. If it is, it means that column was either empty or contained only question marks, in which case the row is considered "improper".

 6. `improper_rows.append(index)` : If an question mark string is found, this line appends (or adds) the index of the current row to the `improper_rows` list.

```python
# dropping all rows with missing values
df = df.drop(df.index[list(set(improper_rows))])
```

```python
# convert label into numerical form
def label_to_numeric(row):
    """Binarize the label."""
    if row["label"] == "ad.":
        return 1
    else:
        return 0

df["label"] = df.apply(label_to_numeric, axis=1)
```

The `label_to_numeric` function is defined to map the "label" from a row in the dataframe to a numerical value. If the label is "ad.", the function will return 1, otherwise, it will return 0. 2.

When `df.apply(label_to_numeric, axis=1)` is called, pandas will apply the `label_to_numeric` function to every row (`axis=1`) in the dataframe `df`.  

The results (i.e., the numeric encodings 1 or 0 corresponding to whether each row's "label" was "ad." or not) are stored back into the original "label" column of the dataframe.

```python
# split the data into training and testing data.
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df)
```

```python
# distribute data into feature arrays and label arrays
y_train = df_train.pop("label").values
y_test = df_test.pop("label").values
X_train = df_train.values
X_test = df_test.values
```

```python
# instantiate a random forest classifier and train (fit) it
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

 we convert our target to numerical form. Then, we train-test split our data in preparation for learning.
